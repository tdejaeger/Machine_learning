# Principal Component Analysis

In this notebook, I will perform principal component analysis (PCA) to perform dimensionality reduction and therefore a better way to visualise the data.

The different steps are:   
-> Standardize the data.     
-> Obtain the Eigenvectors and Eigenvalues from the covariance matrix.    
-> Select k eigenvectors necessary to explain the variance.            
-> Construct the projection matrix from the selected k eigenvectors.      
-> Transform the original dataset via the projection matrix to obtain a k-dimensional feature subspace.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['legend.numpoints']=1
plt.rcParams['xtick.major.size'] = 11
plt.rcParams['xtick.minor.size'] = 5
plt.rcParams['ytick.major.size'] = 11
plt.rcParams['ytick.minor.size'] = 5
plt.rcParams['xtick.minor.visible']=True #See minor tick
plt.rcParams['text.usetex']=True #use Latex
plt.rcParams['axes.linewidth']=2 #width axes
plt.rcParams['axes.labelsize']=25 #
plt.rcParams['ytick.labelsize']=22 #fontsize of tick labels
plt.rcParams['xtick.labelsize']=22 #fontsize of tick labels
plt.rcParams['ytick.direction']='inout' ## direction: in, out, or inout
plt.rcParams['xtick.direction']='inout' ## direction: in, out, or inout
plt.rcParams['xtick.major.top']=True #draw x axis top major ticks
plt.rcParams['xtick.major.bottom']=True #draw x axis bottom major ticks
plt.rcParams['xtick.minor.top']=True ## draw x axis top minor ticks
plt.rcParams['xtick.minor.bottom']=True #draw x axis bottom minor ticks
plt.rcParams['ytick.major.left']=True #draw y axis left major ticks
plt.rcParams['ytick.major.right']=True #draw y axis right major ticks
plt.rcParams['ytick.minor.left']=True ## draw y axis left minor ticks
plt.rcParams['ytick.minor.right']=True #draw y axis right minor ticks
plt.rcParams['font.weight']='bold'
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titlesize']=22
plt.rcParams['figure.titlesize']=22
plt.rcParams['text.latex.preamble']=[r'\boldmath']
```

```python
#First the data: a linearly correlated 2D dataset
X = np.random.uniform(0, 10, 30)[:, np.newaxis]
Y = np.array([i + 2 + np.random.randn() * 1 for i in X])
data = np.hstack((X, Y))

#plot dataset
fig,ax1 = plt.subplots(figsize=(6,6), facecolor='w', edgecolor='k')
ax1.plot(X, Y,marker='o', color='b',alpha=0.7,ms=8,linestyle='None',label=r'\textbf{Data}')
ax1.set_xlabel(r'\textbf{X}')
ax1.set_ylabel(r'\textbf{Y}')
ax1.set_xlim(min(X) - 0.5, max(X) + 0.5)
ax1.set_ylim(min(Y) - 0.5, max(Y) + 0.5)
ax1.legend(loc=0,markerscale=1.0,prop={'size':13},ncol=1)
plt.show()
```

In this figure, we clearly see a linear relationship between the X and Y. Here,we will not try to find the best fit, i.e., predict the y values from the x values but we will use the unsupervised learning problem to attempt to learn about the relationship between the X and Y. Therefore, the PCA find the Principal Axes in the data, and explain how important those axes are in describing the data distribution.

For example, in this case, there are 2 Principal Axes, one along the correlation and one for the dispersion. If the PCA works well, we should also find that the first principal axes is more important than the second.

First we need to standardise the data. It is critical to perform standardization because is quite sensitive regarding the variances of the initial variable. For this purpose, you need to transform the data to comparable scales, for example, you can transform the data onto unit scale (mean = 0 and variance = 1)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca_2d = PCA(n_components=2)
data_std= StandardScaler().fit_transform(data)
X_std,Y_std=data_std[:,0],data_std[:,1]
pca_2d.fit(data_std)
print(pca_2d.explained_variance_ratio_)
print(pca_2d.components_)
print(pca_2d.mean_)
```

```python
#Let's visualise what these numbers are:


fig,ax1 = plt.subplots(figsize=(6,6), facecolor='w', edgecolor='k')
ax1.plot(X_std, Y_std,marker='o', color='b',alpha=0.7,ms=8,linestyle='None')
s_v=4
ax1.plot([pca_2d.mean_[0],pca_2d.mean_[0]+pca_2d.components_[0][0]*s_v*np.sqrt(pca_2d.explained_variance_ratio_[0])],[pca_2d.mean_[1],pca_2d.mean_[1]+pca_2d.components_[0][1]*s_v*np.sqrt(pca_2d.explained_variance_ratio_[0])],'k',lw=3)
ax1.plot([pca_2d.mean_[0],pca_2d.mean_[0]+pca_2d.components_[1][0]*s_v*np.sqrt(pca_2d.explained_variance_ratio_[1])],[pca_2d.mean_[1],pca_2d.mean_[1]+pca_2d.components_[1][1]*s_v*np.sqrt(pca_2d.explained_variance_ratio_[1])],'r',lw=3)


ax1.set_xlabel(r'\textbf{X}')
ax1.set_ylabel(r'\textbf{Y}')
ax1.set_xlim(min(X_std) - 0.5, max(X_std) + 0.5)
ax1.set_ylim(min(Y_std) - 0.5, max(Y_std) + 0.5)
ax1.legend([r'\textbf{Data}', r'\textbf{PC 1}', r'\textbf{PC 2}'],loc=2,markerscale=1.0,prop={'size':15},ncol=1)
plt.show()
```

These two vectors represent the principal axes of the data, and the length of the vector is an indication of how "important" that axis is in describing the distribution of the data. More precisely, it is a measure of the variance of the data when projected onto that axis. 

The explained variance tells you how much information (variance) can be attributed to each of the principal components: here for the variance can be attributed to the first principal component (black) at 97%, so we can easily remove the seconc component without losing information (see below).
 

##### Dimensionality reduction

Now that we have calculated the principal components, we can use them to reduce the number of features in our dataset by projecting each example onto a lower dimensional space (2D to 1D). Here, we will use the eigenvectors and project the example dataset on a 1D space, along the most important principal axis.

```python
pca_2d = PCA(n_components=1)
pca_2d.fit(data_std)
X_pca = pca_2d.transform(data_std)
print("original shape:   ", data_std.shape)
print("transformed shape:", X_pca.shape)
```

Now, X_pca has only 1 dimension!!

```python
X_new = pca_2d.inverse_transform(X_pca)
# Let's see the new data
fig,ax1 = plt.subplots(figsize=(6,6), facecolor='w', edgecolor='k')

#original data (2D)
ax1.plot(X_std, Y_std,marker='o', color='b',alpha=0.7,ms=8,linestyle='None',label=r'\textbf{Data}')
#Projected data (1D)
ax1.plot(X_new[:,0],X_new[:,1],marker='o', color='r',alpha=0.7,ms=8,linestyle='None',label=r'\textbf{Projected data}')
#relationship between original and projected data
for  i in range(X_std.shape[0]):
    ax1.plot([data_std[i,0], X_new[i, 0]], [data_std[i,1], X_new[i,1]], color='black', linestyle='--', linewidth=1.5)

ax1.set_xlabel(r'\textbf{X}')
ax1.set_ylabel(r'\textbf{Y}')
ax1.set_xlim(min(X_std) - 0.5, max(X_std) + 1.0)
ax1.set_ylim(min(Y_std) - 0.5, max(Y_std) + 1.0)
ax1.legend(loc=2,markerscale=1.0,prop={'size':15},ncol=1)
plt.show()
```

In this figure, we clearly see what a PCA dimensionality reduction means: the information along the least important (red in the previous figure) principal axis is removed, leaving only the component of the data with the highest variance.

### 1.2) 4D -> 2D using Iris dataset


```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
#see the data
df.head()
#the iris data have 4 dimensions: sepal length, sepal width, petal length, and pedal width
```

```python
#Seaborn and pairplot are awesome to visualise data
#import seaborn as sns
#sns.set(style="darkgrid")
#sns.set(style="ticks", color_codes=True)
#ax = sns.pairplot(df, hue="target")
#plt.show()
```

We will apply the PCA and see in how many components we can reduce the data.



```python
# Standardisation 
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
# Separating out the features
df_std = df.loc[:, features].values
# Separating out the target
y = df.loc[:,['target']].values
# Standardizing the features
scaler = StandardScaler()
df_std = scaler.fit_transform(df_std)
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(df_std)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2','PC3', 'PC4'])

print(pca.explained_variance_ratio_)
print(pca.components_)
```

```python
fig,ax1 = plt.subplots(figsize=(6,6), facecolor='w', edgecolor='k')
ax1.plot(np.cumsum(pca.explained_variance_ratio_)*100,'ro-',label=r'\textbf{Cumilative}')
height = np.array(pca.explained_variance_ratio_)*100
bars = (r'\textbf{PC1}', r'\textbf{PC2}', r'\textbf{PC3}', r'\textbf{PC4}')
ax1.bar([0,1,2,3], height, color=(0.2, 0.8, 0.6, 0.6),label=r'\textbf{individual}')
plt.xticks([0,1,2,3], bars)
ax1.set_xlabel(r'\textbf{number of components}')
ax1.set_ylabel(r'\textbf{Explained variance in percent}')
ax1.legend(loc=0,markerscale=1.0,prop={'size':15},ncol=1)
plt.show()
```

This figure clearly shows that we can reduce the 4 dimensions of our sample in 2D without losing a lot of information. With 2 components more than 95% of the variance is explained.

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_std)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])
principalDf['target']=y
principalDf.head()
```

```python
#Figure of 2 dimensional data
colors=['r','b','k']
class_y=np.unique(y)
fig,ax1 = plt.subplots(figsize=(6,6), facecolor='w', edgecolor='k')
for j,i in enumerate(class_y):
    x_plot=principalDf['PC1'][principalDf['target']==i]
    y_plot=principalDf['PC2'][principalDf['target']==i]
    ax1.scatter(x_plot,y_plot,s=50,alpha=0.8,color=colors[j],label=r'\textbf{Class: %s}'%i)
ax1.set_xlabel(r'\textbf{PC1}')
ax1.set_ylabel(r'\textbf{PC2}')
ax1.legend(loc=0,markerscale=0.8,prop={'size':13},ncol=1)
plt.show()
```

### 1.3) Face image dataset

```python
#We select 12 faces
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=12)
name_faces=faces.target_names[faces.target]
print(faces.images.shape)
fig, ax = plt.subplots(2, 6, figsize=(18, 6), subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.3, wspace=0.3))
for i, ax in enumerate(ax.flat):
    ax.imshow(faces.data[i].reshape(faces.images.shape[1:3]), cmap='bone')
    ax.set_title(name_faces[i])
plt.show()
```

```python
#use RandomizedPCA faster than the standard PCA estimator
from sklearn.decomposition import PCA as RandomizedPCA
pca = RandomizedPCA(1000)
pca.fit(faces.data)

fig,ax1 = plt.subplots(figsize=(6,6), facecolor='w', edgecolor='k')
ax1.plot(np.cumsum(pca.explained_variance_ratio_)*100,'ro-',label=r'\textbf{Cumilative}')
ax1.set_xlabel(r'\textbf{number of components}')
ax1.set_ylabel(r'\textbf{Explained variance in percent}')
ax1.legend(loc=0,markerscale=1.0,prop={'size':15},ncol=1)
plt.show()
```

```python
#Number of components necessary to explain 95% of the variance 
n_PCA=np.where(np.cumsum(pca.explained_variance_ratio_)*100>95)[0][0]
print('Number of components: %s'%n_PCA)
```

```python
pca = RandomizedPCA(n_PCA)
pca.fit(faces.data)
#We plot the first 18 eigenfaces
fig, ax = plt.subplots(3, 6, figsize=(15, 6), subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.3, wspace=0))
for i, ax in enumerate(ax.flat):
    ax.imshow(pca.components_[i].reshape(faces.images.shape[1:3]), cmap='bone')
    ax.set_title('Eigenfaces %s'%(i+1))

plt.show()
```

```python
# Apply dimensionality reduction to faces.
components = pca.transform(faces.data)
#Transform data back to its original space
projected = pca.inverse_transform(components)

#Let's compare the original image and the projected image
fig, ax = plt.subplots(2, 12, figsize=(18, 6), subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=-0.3, wspace=0.2))

for i,j in enumerate(np.arange(0,12,1)):
    if (i<6):
        ax[0, i+i].imshow(faces.data[i].reshape(faces.images.shape[1:3]), cmap='bone')
        ax[0, i+i].set_title('Original')
        ax[0, i+1+i].imshow(projected[i].reshape(faces.images.shape[1:3]), cmap='bone')
        ax[0, i+1+i].set_title('Projec')
    else:
        ax[1, i+i-12].imshow(faces.data[i].reshape(faces.images.shape[1:3]), cmap='bone')
        ax[1, i+i-12].set_title('Original')
        ax[1, i+i-11].imshow(projected[i].reshape(faces.images.shape[1:3]), cmap='bone')
        ax[1, i+i-11].set_title('Projec')
            
plt.show()
```

In the figure above, we show the original images and the images from just ~180 of the almost 3000 initial features.
Even if we reduced the dimensionality of data by a factor 20, the projected images contain enough information to reconize the individual image. 

What this means is that our classification algorithm needs to be trained on 150-dimensional data rather than 3,000-dimensional data, which depending on the particular algorithm we choose, can lead to a much more efficient classification.
